```python
####analyse the data and visualization
```

```python
# Calculate the cancellation percentage
can_per = df['is_canceled'].value_counts(normalize=True)
print(can_per)

# Plot the bar chart
plt.figure(figsize=(5, 4))
plt.title('Reservation Status')
plt.bar(['Not Canceled', 'Canceled'], df['is_canceled'].value_counts())
plt.show()
```

```python
plt.figure(figsize=(8, 4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette="Blues")
plt.title("Reservation Status Hotel Wise")
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
City_hotel = df[df['hotel'] == 'City Hotel']
City_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
City_hotel = City_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20, 8))
plt.title('ADR in city and resort', fontsize=30)  
plt.plot(resort_hotel.index, resort_hotel['adr'], label='resort_hotel')  
plt.plot(City_hotel.index, City_hotel['adr'], label='City_hotel')  
plt.legend()  # Added legend to display labels
plt.show()
```

```python
# the price of the city hotell is less than the resort hotel and the price of the hottel are increase in the weackens 
```

```python
df['month'] = df['reservation_status_date'].dt.month 
plt.figure(figsize=(16, 8))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df) 
plt.show()
```

```python
# Assuming 'month' is a column in your DataFrame
df['month'] = df['reservation_status_date'].dt.month

# Create a bar plot for ADR per month for canceled reservations
plt.figure(figsize=(20, 8))
plt.title('ADR per month', fontsize=30)

# Group by month and calculate the sum ADR for canceled reservations
adr_per_month = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()

# Use seaborn barplot for visualization
sns.barplot(x='month', y='adr', data=adr_per_month)
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]
top_1_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(20, 8))
plt.title('top_1_country with reservation canceled')
plt.pie(top_1_country,autopct = '%.2f' , labels = top_1_country.index )
plt.show()
```

```python
# Assuming you have 'cancelled_data' DataFrame
cancelled_data = df[df['is_canceled'] == 1]

cancelled_data_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_data_adr.reset_index(inplace=True)
cancelled_data_adr['reservation_status_date'] = pd.to_datetime(cancelled_data_adr['reservation_status_date'])
cancelled_data_adr.sort_values('reservation_status_date', inplace=True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()  # Convert to DataFrame
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr['reservation_status_date'] = pd.to_datetime(not_cancelled_df_adr['reservation_status_date'])
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate')

# Plot ADR for not cancelled reservations
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')
plt.plot(cancelled_data_adr['reservation_status_date'], cancelled_data_adr['adr'], label='Cancelled')

plt.legend()
plt.show()
```

```python
#for removing the inconsistsncy in the data
cancelled_data_adr = cancelled_data_adr[(cancelled_data_adr['reservation_status_date'] > '2016' ) & (cancelled_data_adr['reservation_status_date'] < '2017-09' )]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date'] > '2016' ) & (not_cancelled_df_adr['reservation_status_date'] < '2017-09' )]
```
